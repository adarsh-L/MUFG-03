```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset from the data folder
df = pd.read_csv("data/manufacturing_dataset_1000_samples.csv")

# Summary statistics
display(df.describe())

# Correlation heatmap: use only numeric columns!
num_cols = df.select_dtypes(include='number').columns
plt.figure(figsize=(10, 6))
sns.heatmap(df[num_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()

# Boxplots to detect outliers
for col in num_cols:
    plt.figure()
    sns.boxplot(x=df[col], color='gold')
    plt.title(f"Boxplot for {col}")
    plt.show()

# Histograms of all numeric features
df[num_cols].hist(figsize=(12,8), bins=20)
plt.tight_layout()
plt.show()

# Pairplot (relationships) - limit columns if dataset is large
cols = num_cols[:10]
sns.pairplot(df[cols].sample(min(200, len(df))))
plt.show()
```
